```python
import cv2
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(15, 15))

rows = 2
columns = 1

image1 = cv2.imread('/content/drive/MyDrive/blur_images/outputs/saved_images_v3/val_deblurred0.jpg')
image2 = cv2.imread('/content/drive/MyDrive/blur_images/outputs/saved_images_v3/val_deblurred39.jpg')

# converting to RGB
rgb_image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
rgb_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

fig.add_subplot(rows, columns, 1)
plt.imshow(rgb_image1)
plt.axis('off')
plt.title('Blurred Image')

fig.add_subplot(rows, columns, 2)
plt.imshow(rgb_image2)
plt.axis('off')
plt.title('Deblurred Image')

plt.show()
```
